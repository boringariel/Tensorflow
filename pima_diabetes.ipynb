```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
df_dia = pd.read_csv('./diabetes.csv')
```

```python
df_dia.head()
# 임신 횟수, 식후 2시간 경과 혈당 수치, 혈압 등의 8개 항목과 당뇨병 여부를 나타내고 있다
```

```python
sb.clustermap(df_dia,z_score=0)
# 각 요소별 연관성을 찾기 위해 clustermap을 작성함
```

```python
model = Sequential()
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
col = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
```

```python
x = df_dia[col].values
```

```python
y = df_dia['Outcome'].values
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x,y,epochs=200,batch_size=10)
```
